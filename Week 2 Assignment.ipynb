This notebook explains set comprehensions in Python. Set comprehensions help create sets quickly, remove duplicates, and filter values.
Set: A set is a collection that stores unique values (no duplicates). Sets are useful when I want to remove duplicates fast

```python
nums = [1, 2, 2, 3, 4, 4, 4]
unique_nums = {n for n in nums}
unique_nums
```

## Set comprehension syntax

General form:

{ expression for item in iterable }

It looks like a list comprehension, but it creates a set instead of a list.

**Note:** Sets do not keep order, so results may show in a different order.
Also `{}` makes an empty dictionary, not an empty set. Use `set()` for empty set.

```python
empty_set = set()
empty_set
```

```python
nums = [1, 2, 3, 4, 5]
squares = {n*n for n in nums}
squares
```

## Using an if condition (filter)

I can filter values inside a set comprehension using `if`.

```python
nums = range(1, 11)
evens = {n for n in nums if n % 2 == 0}
evens
```

```python
words = ["Apple", "apple", "BANANA", "banana", "Orange", "ORANGE"]
unique_words = {w.lower() for w in words}
unique_words
```

## Quick practice

Create a set of numbers greater than 5 from this list:

[1, 2, 5, 6, 6, 7, 10]

My solution is below.

```python
nums = [1, 2, 5, 6, 6, 7, 10]
greater_than_5 = {n for n in nums if n > 5}
greater_than_5
```

## Summary
- Set comprehensions create sets in one line.
- Sets store **unique** values (no duplicates).
- I can filter items using `if`.
- Useful for cleaning data and removing duplicates.
